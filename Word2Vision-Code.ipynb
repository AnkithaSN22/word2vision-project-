```python
from diffusers import DiffusionPipeline, StableDiffusionPipeline
import torch

# Option 1: Using DiffusionPipeline
pipe = DiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float16,
    use_auth_token=True
)
pipe.to("cuda")
```

```python
prompt = "a detailed portrait of naruto from the anime naruto, focusing only on his face"
negative_prompt = "blurring, including other background elements"

images = pipe(prompt=prompt, negative_prompt=negative_prompt).images[0]
images.save("naruto.png")

# Display the image
import matplotlib.pyplot as plt
import numpy as np

plt.imshow(images)
plt.axis("off")
plt.show()
```
